```python
df_copy= df
```

```python
df_copy['sex'] = df.sex.replace({1: "Male", 0: "Female"})
df_copy['anaemia'] = df.anaemia.replace({1: "Yes", 0: "No"})
df_copy['diabetes'] = df.diabetes.replace({1: "Yes", 0: "No"})
df_copy['high_blood_pressure'] = df.high_blood_pressure.replace({1: "Yes", 0: "No"})
df_copy['smoking'] = df.smoking.replace({1: "Yes", 0: "No"})

df_copy['DEATH_EVENT'] = df.DEATH_EVENT.replace({1: "Died", 0: "Alive"})
```
